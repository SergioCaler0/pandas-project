```python
import pandas as pd
import src.limpieza_texto as lt
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
data = pd.read_csv("data/clean-sharks.csv",encoding = "ISO-8859-1")
```

```python
data=data.drop(['Unnamed: 0', 'Date'],axis=1)
```

```python
data.head()
```

```python
sns.countplot(y=data.Country)
```

```python
sns.histplot(y='Country', hue='Fatal (Y/N)', data= data, palette='RdYlBu')
```

```python
sns.histplot(y='species', hue='Fatal (Y/N)', data= data)
```

Casos de ataques de tiburon por paises

```python
plot = sns.kdeplot(x=data.Age)
plot.axvline(x=data.Age.mean(), c="red", label="mean")
plot.axvline(x=data.Age.dropna().median(), c="green", label="median")
plt.legend()
```

Grafico de como aumentan los casos con el avance de los aos

```python
sns.kdeplot(x=data.Year, hue=data.species)
```

```python
sns.boxplot(x="Year", data=data)
```
